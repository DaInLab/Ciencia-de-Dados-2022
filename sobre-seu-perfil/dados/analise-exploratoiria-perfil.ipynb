No Python, para utilizar determinadas funções, que não estão nos recursos básicos é preciso "carregar" a biblioteca no programa. 
O comando que realiza esta função é chamado de "import".
Neste caso, utilizaremos a "biblioteca" pandas.  
Pandas é uma biblioteca que ajuda a lidar com dados, realizar operações e analisá-los com plotagens e gráficos.

```python
# Carregando a biblioteca pandas
import pandas as pd
```

```python
# Lendo a planilha e carregando-a no Python como "dataframe". 
# Para ler arquivos .xlsx usando pandas, podemos usar a função read_excel(). 
# Esta função lê um arquivo excel em um pandas Dataframe. 
```

```python
meu_dataframe = pd.read_excel("D:\Github\Ciencia-de-Dados-2022\sobre-seu-perfil\dados\sobre-seu-perfil-aed.xlsx", 
                             sheet_name="Respostas")
```

```python
#O dataframe é usado para armazenar as tabelas de dados em Python. 
#É uma tabela ou uma estrutura bidimensional do tipo matriz em que cada coluna contém os valores de uma variável
#e cada linha contém um conjunto de valores de cada coluna.
#É uma lista de vetores de igual comprimento.
```

### Alguns gráficos básicos
- Tipo pizza

#### Sistema operacinal utilizado

```python
labels = 'Windows', 'MacOs'
```

```python
meu_dataframe.sistema_operacional.value_counts()
```

```python
sizes = [13, 5]
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(15,8)) # tamanho da pizza
plt.legend(title='Sistemas Operacionais Utilizados pelos Alunos',loc=1) # Título do gráfico
plt.pie(sizes, labels=labels,  autopct='%1.1f%%', shadow=True, startangle=90, colors=['gray', 'orange'])
plt.show()
```

```python
# Versão 2
```

```python
fig1, axl = plt.subplots()
axl.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90, colors=['blue', 'yellow'])
axl.axis('equal')
axl.set_title('Sistemas Operacionais Utilizados pelos Alunos')
plt.show()
```
